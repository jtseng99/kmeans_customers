```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
cust_df = pd.read_csv("Cust_Segmentation.csv")
cust_df
```

```python
cust_df.columns.values
```

```python
#drop customer id and address
cust_df = cust_df.reindex(columns=['Age', 'Edu', 'Years Employed', 'Income', 'Card Debt', 'Other Debt', 'Defaulted', 'DebtIncomeRatio'])
```

```python
cust_df
```

```python
#filter NaN
cust_df[cust_df.isna().any(axis=1)]
```

```python
cust_df.fillna(0,inplace=True)
cust_df
```

```python
#check again
cust_df[cust_df.isna().any(axis=1)]
```

```python
#normalize to standard deviations
X = cust_df.values
X
```

```python
X = StandardScaler().fit_transform(X)
X
```

```python
clusterNum = 3
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
```

```python
k_means.labels_
```

```python
k3_cust_df = cust_df
k3_cust_df
```

```python
k3_cust_df['Cluster'] = k_means.labels_
k3_cust_df
```

```python
k3_cust_df.groupby('Cluster').mean()
```

```python
#k = 5
clusterNum = 5
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
k5_cust_df = cust_df
k5_cust_df['Cluster'] = k_means.labels_
k5_cust_df.groupby('Cluster').mean()
```

```python
#k = 7
clusterNum = 7
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
k7_cust_df = cust_df
k7_cust_df['Cluster'] = k_means.labels_
k7_cust_df.groupby('Cluster').mean()
```

```python
#only use age, edu, income
X = cust_df.reindex(columns=["Age","Edu","Income"]).values
X = StandardScaler().fit_transform(X)
clusterNum = 3
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
k3_cust_df2 = cust_df.reindex(columns=["Age","Edu","Income"])
k3_cust_df2['Cluster'] = k_means.labels_
k3_cust_df2.groupby('Cluster').mean()
```
